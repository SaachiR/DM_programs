```python
#Title: Iris Plants Database
'''
 Number of Instances: 150 (50 in each of three classes)
 Number of Attributes: 4 numeric, predictive attributes and the class

Attribute Information:
1. sepal length in cm
2. sepal width in cm
3. petal length in cm
4. petal width in cm
5. class: 
 -- Iris Setosa
 -- Iris Versicolour
 -- Iris Virginica

'''
import numpy as np
import csv
import random
with open(r'C:\Users\DELL\Desktop\iris.txt') as csvfile:
    #To read data from CSV files, here is the reader function to generate a reader object.
#The reader function is developed to take each row of the file and make a list of all columns.
    instances = csv.reader(csvfile)
    data= list(instances)
print(data)
#convert data from list into an array
data1= np.asarray(data)
print('data :')
print(data1)
X = data1[:100,:-1]
print('X :')
print("Shape of array:\n", np.shape(X)) #shape function prints the dimension of the array
print(X)
res = X.astype(np.float) 
print("Covarinace matrix of x:\n", np.cov(res.T)) # cov fuvtion finds the covvariance and T is the transpose
cov_matrix=np.cov(res.T)
inv_cov = np.linalg.inv(cov_matrix) #find the inverse of the covariance matrix
print(inv_cov)
avg=[]
for i in range(4):
    avg.append(np.mean(res[:,i]))
print("mean :\n", avg)
A = (res - avg)
B=A.T
```

```python
C = np.dot(A,inv_cov) #formula for mahalanobis distance 
D = np.dot(C,B)
print(D)
print(D.shape)
```

```python
#find the outliers
deviation= np.std(D)
center = np.mean(D)
U_range = center + 5*deviation
L_range= center - 5*deviation
outliers = []
for i in range(100):
    for j in range(100):
        if i==j:
            if (D[i][j]>= U_range) or (D[i][j] <= L_range):
                outliers.append(i)
print('Index of the potential outliers : ')
print(np.unique(outliers))
```
