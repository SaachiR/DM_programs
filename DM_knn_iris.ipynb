```python
import csv
import random
with open(r'C:\Users\DELL\Desktop\iris.txt') as csvfile:
    instances = csv.reader(csvfile)
    data= list(instances)
    
# for loop converts the character instances to float in the dataset so that the arthmetic operations can be applied.

for x in range(len(data)-1):
    for y in range(4):
        data[x][y]= float(data[x][y])
    
# random function shuffles the dataset and mix the class randomly everytime the code executes

random.shuffle(data)
```

```python
# splitting the dataset into training and testing part

trainingset=[]
testingset=[]
split=100
for x in range(len(data)-1):
    if x<split:
        trainingset.append(data[x])
    else:
        testingset.append(data[x])
```

```python

print('train_instances: '+ repr(len(trainingset)))
print('test_instances: ' + repr(len(testingset)))

```

```python
import math
def cal_dis(para1, para2, dim):
    '''
    Objective: To calculate the euclidean diatance between the parameter 1 and 2
    Input: para1, para2 and the dimension(no of features in the dataset)
    Return Value: squareroot of distance
    '''
    distance=0
    for x in range(dim):
        distance += pow(para1[x]- para2[x],2)
    return math.sqrt(distance)
        
```

```python
import operator 
def knn(train, test, k):
    '''
    Objective: To predict the class of the test data instance
    Input: training set, test instance and k= no. of nearest neighbours
    Output: prints the predicted value of the test data instance
    '''
    distances=[]
    print('test_instance : ' + repr(test))
    dim= len(test)-2
    for x in range(len(train)-1):
        print('train_instance : ' + repr(train[x]))
        d=cal_dis(test, train[x], dim)
        distances.append((x, d))
        
        # list distances contains the euclidean distance of the test data instance from the training instances.
        # x= index of the training instance
        # d= the euclidean distance of the test data instance from the training instance x.
        
    print('distances :')
    print(distances)
    distances.sort(key=operator.itemgetter(1))
    kneighbours=[]
    
    # list kneighbours contains the k nearest neighbours to the test instance
    
    for x in range(k):
        kneighbours.append(distances[x][0])
    print('kneighbours : ')
    print(kneighbours)
    count=[0,0,0]
    #count list contains the votes of each class
    
    for x in kneighbours:
        kclass= train[x][-1]
        if kclass == 'Iris-setosa':
            count[0]+=1
        elif kclass=='Iris-versicolor':
            count[1]+= 1
        else:
            count[2]+=1
    count_index= count.index(max(count))
    print('predicted class : ')
    if count_index == 0:
        print('Iris-setosa')
    elif count_index==1:
        print('Iris-versicolor')
    else:
        print('Iris-virginica')



        
```
